```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("products_asos.csv", on_bad_lines='skip')
```

```python
df.head(10)
```

```python
df['price'] = pd.to_numeric(df['price'], errors='coerce')
df = df.dropna(subset=['price'])
```

```python
print(f'Data Loaded : {len(df)} rows')
df.head(5)
```

# text splitting for consistency brand in column

```python
df['description'] = df['description'].astype(str)
df['description']
```

```python
def get_brand(text):
    if 'by ' in text:
        try:
            return text.split('by ')[1].split(' ')[0]
        except:
            return 'Unknown'
    return 'Unknown'
```

```python
df["brand_raw"] = df["description"].apply(get_brand)
```

```python
df
```

```python
brand_category_count = df["brand_raw"].value_counts()
brand_category_count
```

# Redefined category of brand use mapping

```python
brand_map = {
    'New': 'New Look',
    'River': 'River Island',
    'Miss': 'Miss Selfridge',
    'TopshopWelcome': 'Topshop'
}

df['Brand'] = df['brand_raw'].map(brand_map).fillna(df['brand_raw'])
```

```python
brand_count = df['Brand'].value_counts()
valid_brand = brand_count[brand_count > 5].index
df_clean = df[df['Brand'].isin(valid_brand)].copy()

print(df_clean["Brand"].value_counts().head(5))
```

```python
df_clean["size"].value_counts()
```

# function to analyze stockouts

```python
def calculate_phantom_revenue(size_str):
    if not isinstance(size_str, str):
        return 0, 0.0
    # split "UK-6, UK-8 - Out of stock" into list
    sizes = size_str.split(',')
    total_size = len(sizes)
    # Count how many items are out of stock
    out_of_stocks = size_str.count('Out of stock')
    # Calculate Rate
    rate = out_of_stocks / total_size if total_size > 0 else 0.0
    return out_of_stocks, rate

metrics = df_clean['size'].apply(lambda x: calculate_phantom_revenue(x))
df_clean["Stockout_Count"] = [x[0] for x in metrics]
df_clean["Stockout_Rate"] = [x[1] for x in metrics]
```

```python
df_clean[["Stockout_Count", "Stockout_Rate"]].head(9)
```

```python
# missing revenue cause of Out of stock
df_clean["Lost"] = df_clean["price"] * df_clean["Stockout_Count"]
```

```python
cols = ["Brand", "name", "price", "Stockout_Count", "Lost"]
print(df_clean.sort_values(by='Lost', ascending=False).head(5)[cols])
```

```python
df_clean.info()
```

```python
brand_strat = df_clean.groupby('Brand').agg({
    'price': 'mean',
    'Stockout_Rate': 'mean',
    'Lost': 'sum',
    'name': 'count'
}).reset_index()
brand_strat
```

```python
brand_strategy = brand_strat[brand_strat["name"] > 10]
plt.figure(figsize=(12, 8))
sns.scatterplot(
    data=brand_strat,
    x='price',
    y='Stockout_Rate',
    size='Lost',
    sizes=(50, 500),
    alpha=0.7,
    palette='viridis'
)

winner = brand_strategy[
    (brand_strategy['price'] > 40) &
    (brand_strategy['Stockout_Rate'] > 0.4)
]

for i in range(len(winner)):
    plt.text(
        winner.iloc[i]['price'] + 1,
        winner.iloc[i]['Stockout_Rate'],
        winner.iloc[i]['Brand'],
        # fontsize=12,
        # color='black',
        # weight='bold'
    )

plt.title('Brand Strategy Analysis')
plt.xlabel('Average Price')
plt.ylabel('Stockout Rate')
plt.axvline(x=40, color='red', linestyle='--')
plt.axvline(y=0.4, color='red', linestyle='--')
plt.show()
```
